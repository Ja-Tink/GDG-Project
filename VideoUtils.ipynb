```python
import cv2
from matplotlib import pyplot as plt
```

```python
video_path = 'Videos/NYU Tour.mp4'
```

```python
'''
Function to get <num> evenly spaces frames from <video_path>
'''
def extractFrames(video_path, num):
    cap = cv2.VideoCapture(video_path)

    outlist = []
    
    # get total number of frames
    totalFrames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    
    frame_indices = [(totalFrames // 10) * multiplier for multiplier in range(0, num)]

    print(frame_indices)
    
    for frame_number in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        ret, frame = cap.read()

        outlist.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

        '''
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB), interpolation='nearest')
        plt.show()
        '''
    
    cap.release()

    return outlist
```

```python
frameList = extractFrames(video_path, 10)

for i in frameList:
    plt.imshow(i, interpolation = 'nearest')
    plt.show()
```
